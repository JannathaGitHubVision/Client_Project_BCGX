```python
price_df = pd.read_csv('price_data.csv')
price_df["price_date"] = pd.to_datetime(price_df["price_date"], format='%Y-%m-%d')
price_df.head()
```

```python
# Group off-peak prices by companies and month
monthly_price_by_id = price_df.groupby(['id', 'price_date']).agg({'price_off_peak_var': 'mean', 'price_off_peak_fix': 'mean'}).reset_index()

# Get january and december prices
jan_prices = monthly_price_by_id.groupby('id').first().reset_index()
dec_prices = monthly_price_by_id.groupby('id').last().reset_index()

# Calculate the difference
diff = pd.merge(dec_prices.rename(columns={'price_off_peak_var': 'dec_1', 'price_off_peak_fix': 'dec_2'}), jan_prices.drop(columns='price_date'), on='id')
diff['offpeak_diff_dec_january_energy'] = diff['dec_1'] - diff['price_off_peak_var']
diff['offpeak_diff_dec_january_power'] = diff['dec_2'] - diff['price_off_peak_fix']
diff = diff[['id', 'offpeak_diff_dec_january_energy','offpeak_diff_dec_january_power']]
diff.head()
```

```python
price_df.info()
```

```python
# Aggregate price data by company ID (get mean prices across all dates)
price_aggregated = price_df.groupby('id').agg({
    'price_off_peak_var': 'mean',
    'price_peak_var': 'mean', 
    'price_mid_peak_var': 'mean',
    'price_off_peak_fix': 'mean',
    'price_peak_fix': 'mean',
    'price_mid_peak_fix': 'mean'
}).reset_index()

# Merge with featureEng dataset
featureEng = pd.merge(featureEng, price_aggregated, on='id', how='left')

# Check the combined dataset
featureEng.info()
```

```python
single_val_cols = []

for column in featureEng.columns:
    if featureEng[column].nunique() == 1:
        single_val_cols.append(column)

print('COlumns with only 1 unique value')
print(single_val_cols)

numeric_columns = featureEng.select_dtypes(include=[np.number]).columns
correlation_matrix = featureEng[numeric_columns].corr()

high_corr_pairs = []

for i in range(len(correlation_matrix.columns)):
    for j in range(i+1,len(correlation_matrix.columns)):
        if abs(correlation_matrix.iloc[i,j]) > 0.95:
                       high_corr_pairs.append((correlation_matrix.columns[i], correlation_matrix.columns[j], correlation_matrix.iloc[i, j]))


print("\nHighly correlated column pairs (>0.95):")
for pair in high_corr_pairs:
    print(f"{pair[0]} - {pair[1]}: {pair[2]:.3f}")

# Remove columns
columns_to_remove = single_val_cols + ['id']  # Remove id as it's just an identifier

# Remove one column from each highly correlated pair
for pair in high_corr_pairs:
    if pair[1] not in columns_to_remove:  # Keep the first, remove the second
        columns_to_remove.append(pair[1])

print(f"\nColumns to remove: {columns_to_remove}")

# Remove the columns
featureEng_cleaned = featureEng.drop(columns=columns_to_remove)

print(f"\nOriginal shape: {featureEng.shape}")
print(f"Cleaned shape: {featureEng_cleaned.shape}")
print(f"Removed {len(columns_to_remove)} columns")
```

```python
# Create new combined features
featureEng_enhanced = featureEng_cleaned.copy()

# 1. Price volatility features
featureEng_enhanced['price_volatility_6m'] = (featureEng_enhanced['var_6m_price_off_peak_var'] + 
                                            featureEng_enhanced['var_6m_price_peak_var'] + 
                                            featureEng_enhanced['var_6m_price_mid_peak_var']) / 3

featureEng_enhanced['price_volatility_year'] = (featureEng_enhanced['var_year_price_off_peak_var'] + 
                                              featureEng_enhanced['var_year_price_peak_var'] + 
                                              featureEng_enhanced['var_year_price_mid_peak_var']) / 3

# 2. Consumption efficiency
featureEng_enhanced['consumption_efficiency'] = featureEng_enhanced['forecast_cons_year'] / featureEng_enhanced['cons_last_month']

# 3. Price spread (difference between peak and off-peak)
featureEng_enhanced['price_spread'] = featureEng_enhanced['forecast_price_energy_peak'] - featureEng_enhanced['forecast_price_energy_off_peak']

# 4. Total price exposure
featureEng_enhanced['total_price_exposure'] = (featureEng_enhanced['forecast_price_energy_peak'] + 
                                             featureEng_enhanced['forecast_price_energy_off_peak'] + 
                                             featureEng_enhanced['forecast_price_pow_off_peak'])

# 5. Price change momentum (6m vs year)
featureEng_enhanced['price_momentum_off_peak'] = featureEng_enhanced['var_6m_price_off_peak_var'] - featureEng_enhanced['var_year_price_off_peak_var']
featureEng_enhanced['price_momentum_peak'] = featureEng_enhanced['var_6m_price_peak_var'] - featureEng_enhanced['var_year_price_peak_var']

# 6. Customer tenure in months (assuming current year is 2016 based on typical dataset)
featureEng_enhanced['tenure_months'] = featureEng_enhanced['num_years_antig'] * 12

# 7. Seasonal activation (create season from activation month)
featureEng_enhanced['activation_season'] = pd.cut(featureEng_enhanced['date_activ_month'], 
                                                 bins=[0, 3, 6, 9, 12], 
                                                 labels=['Winter', 'Spring', 'Summer', 'Fall'])

# 8. Days of week pattern (weekday vs weekend)
featureEng_enhanced['activ_is_weekend'] = (featureEng_enhanced['date_activ_dayofweek'] >= 5).astype(int)
featureEng_enhanced['end_is_weekend'] = (featureEng_enhanced['date_end_dayofweek'] >= 5).astype(int)

# 9. Product count per tenure (product intensity)
featureEng_enhanced['products_per_year'] = featureEng_enhanced['nb_prod_act'] / (featureEng_enhanced['num_years_antig'] + 1)

print("New features created:")
new_features = ['price_volatility_6m', 'price_volatility_year', 'consumption_efficiency', 
                'price_spread', 'total_price_exposure', 'price_momentum_off_peak', 
                'price_momentum_peak', 'tenure_months', 'activation_season', 
                'activ_is_weekend', 'end_is_weekend', 'products_per_year']

for feature in new_features:
    print(f"- {feature}")

print(f"\nOriginal shape: {featureEng_cleaned.shape}")
print(f"Enhanced shape: {featureEng_enhanced.shape}")
print(f"Added {len(new_features)} new features")
```
