## 3. Data visualization

Now let's dive a bit deeper into the dataframes

```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    # Labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )
```

### Consumption

Let's see the distribution of the consumption in the last year and month. Since the consumption data is univariate, let's use histograms to visualize their distribution.

```python
consumption = client_df[['id', 'cons_12m', 'cons_gas_12m', 'cons_last_month', 'imp_cons', 'has_gas', 'churn']]
```

```python
def plot_distribution(dataframe, column, ax, bins_=50, show_stats=True):
    """
    Plot variable distribution in a stacked histogram of churned or retained company
    """
    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')


    if show_stats:
        retention_stats = temp['Retention'].describe()
        churn_stats = temp['Churn'].describe()
        stats_text = (
           f"Retention: {len(temp['Retention']):,} customers\n"
            f"   Mean: {retention_stats['mean']:,.2f}\n"
            f"   Median: {retention_stats['50%']:,.2f}\n"
            f"   Max: {retention_stats['max']:,.2f}\n\n"
            f"Churn: {len(temp['Churn']):,} customers\n"
            f"   Mean: {churn_stats['mean']:,.2f}\n"
            f"   Median: {churn_stats['50%']:,.2f}\n"
            f"   Max: {churn_stats['max']:,.2f}"
                   )

          # Add text box with statistics
        props = dict(boxstyle='round', facecolor='white', alpha=0.7)
        ax.text(0.05, 0.95, stats_text, transform=ax.transAxes, fontsize=10,
                verticalalignment='top', bbox=props)
```

```python
fig, axs = plt.subplots(nrows=4, figsize=(18, 25))

plot_distribution(consumption, 'cons_12m', axs[0])
plot_distribution(consumption[consumption['has_gas'] == 't'], 'cons_gas_12m', axs[1])
plot_distribution(consumption, 'cons_last_month', axs[2])
plot_distribution(consumption, 'imp_cons', axs[3])

plt.tight_layout()
plt.show()
```

Clearly, the consumption data is highly positively skewed, presenting a very long right-tail towards the higher values of the distribution. The values on the higher and lower end of the distribution are likely to be outliers. We can use a standard plot to visualise the outliers in more detail. A boxplot is a standardized way of displaying the distribution based on a five number summary:
- Minimum
- First quartile (Q1)
- Median
- Third quartile (Q3)
- Maximum

It can reveal outliers and what their values are. It can also tell us if our data is symmetrical, how tightly our data is grouped and if/how our data is skewed.

```python
from matplotlib.ticker import ScalarFormatter

fig, axs = plt.subplots(nrows=4, figsize=(18,25))

# Plot boxplots
sns.boxplot(consumption["cons_12m"], ax=axs[0])
sns.boxplot(consumption[consumption["has_gas"] == "t"]["cons_gas_12m"], ax=axs[1])
sns.boxplot(consumption["cons_last_month"], ax=axs[2])
sns.boxplot(consumption["imp_cons"], ax=axs[3])

# Remove scientific notation on y-axis
for ax in axs:
    ax.yaxis.set_major_formatter(ScalarFormatter())
# Set x-axis limits for the first three plots
axs[0].set_xlim(-200000, 2000000)
axs[1].set_xlim(-200000, 2000000)
axs[2].set_xlim(-20000, 100000)
plt.show()
```

# Key Insights from Energy Customer Churn Analysis

## Overall Churn Rate
- About 10% of customers have churned, which is fairly typical for utility companies.
- Understanding the factors driving this churn can help develop effective retention strategies.

## Customer Consumption Patterns
- Higher consumption customers tend to be more loyal (both electricity and gas).
- The average consumption for retained customers is significantly higher than churned customers.
- This suggests the company should focus retention efforts on smaller consumers who appear more likely to switch.

## Forecasted Values and Pricing
- Customers who churned were expected to pay higher meter rental fees (72 vs 62 units).
- Peak energy rates were slightly higher for churned customers.
- Higher forecasted discounts for churned customers (1.23% vs 0.94%) weren't enough to prevent churn.
- Price sensitivity appears to be a key driver of customer decisions to leave.

## Sales Channel Impact
- Churn rates vary significantly across different sales channels.
- Customers with missing sales channel data show moderate churn (7.6%).
- This indicates acquisition method influences long-term customer retention.

## Recommendations
- Develop targeted retention strategies for smaller energy consumers.
- Review pricing structure, particularly meter rental fees and peak rates.
- Consider more substantial discounts or loyalty programs for price-sensitive customers.
- Further analyze high-value customers to understand factors driving their loyalty.
- Address issues in sales channels with higher churn rates.

The data shows that pricing factors and consumption patterns play important roles in customer retention. A segmented approach addressing the specific needs of different customer groups will likely be more effective than a one-size-fits-all retention strategy.

Similarly to the consumption plots, we can observe that a lot of the variables are highly positively skewed, creating a very long tail for the higher values. We will make some transformations during the next exercise to correct for this skewness.
