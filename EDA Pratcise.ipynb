## **What We're Trying to Do: Customer Churn Analysis**

Think of this like running a business (like an electricity company) and you want to understand:

### **The Big Question:**
*"Why do some customers leave us (churn) while others stay?"*

### **What We're Doing Step by Step:**

1. **üîç Look at the Data**
   - We have customer information (age, usage, etc.)
   - We have price history (what each customer paid over time)
   - We know who left (churn = 1) and who stayed (churn = 0)

2. **üîó Connect the Information**
   - Instead of having messy data with multiple price records per customer
   - We create ONE clean row per customer with their recent price AND average price

3. **üìä Explore Patterns**
   - "Do customers who pay higher prices leave more often?"
   - "What's the overall rate of customers leaving?"
   - "Are there price patterns that predict who will leave?"

4. **üîç Test Our Suspicions**
   - Check if high prices = more likely to leave
   - Look at different price ranges and see churn rates
   - See if recent price increases made people leave

5. **üõ†Ô∏è Create New Insights**
   - Calculate things like "price volatility" (how much prices jump around)
   - Find "price trends" (are prices going up or down for each customer?)
   - Create features that might predict churn better

### **The Goal:**
Understand what makes customers leave so the business can:
- Keep prices fair
- Identify at-risk customers
- Take action to prevent churn

**Simple analogy:** It's like being a detective trying to figure out why some restaurant customers stop coming back - is it the prices, the service, or something else?

```python
# Check the actual columns to confirm
print("Price Data columns:", priceData.columns.tolist())
print("Client Data columns:", clientData.columns.tolist())

# Since you ahave price_date column, we can get the most recent price per customer
# Let's use price_off_peak_var as the main price column (you can change this based on your preference)

# Most recent price per customer using price_date
most_recent = priceData.loc[priceData.groupby('id')['price_date'].idxmax()][['id', 'price_off_peak_var', 'price_peak_var', 'price_mid_peak_var']]
most_recent.rename(columns={
    'price_off_peak_var': 'recent_price_off_peak', 
    'price_peak_var': 'recent_price_peak',
    'price_mid_peak_var': 'recent_price_mid_peak'
}, inplace=True)

# Average prices per customer
avg_prices = priceData.groupby('id')[['price_off_peak_var', 'price_peak_var', 'price_mid_peak_var']].mean().reset_index()
avg_prices.rename(columns={
    'price_off_peak_var': 'avg_price_off_peak',
    'price_peak_var': 'avg_price_peak', 
    'price_mid_peak_var': 'avg_price_mid_peak'
}, inplace=True)

# Merge all together
merged_data_complete = pd.merge(clientData, most_recent, on='id', how='inner')
merged_data_complete = pd.merge(merged_data_complete, avg_prices, on='id', how='inner')

print(f"Final merged dataset shape: {merged_data_complete.shape}")
merged_data_complete.head()
```

## **Step 3: Deep Dive Analysis - Getting the Full Picture**

We're taking a **magnifying glass** to each price type to understand EXACTLY how they differ between customers who left vs those who stayed.

### **What It Does:**

**üîç For Each Price Type (Off-peak, Peak, Mid-peak), it:**

1. **Separates customers into 2 groups:**
   - Blue group = Customers who stayed (churn = 0)
   - Red group = Customers who left (churn = 1)

2. **Calculates detailed statistics for each group:**
   - How many customers in each group
   - Average price (mean)
   - Middle price (median) 
   - How spread out the prices are (std dev)
   - Lowest and highest prices

3. **Creates visual charts:**
   - Boxplots = Shows price ranges side-by-side
   - Histograms = Shows how many customers pay each price level

### **What We Discovered:**

**üìä The Key Findings:**

**Off-Peak Prices:** 
- Customers who left actually pay **0.45% LESS** (tiny difference)
- **Conclusion:** Off-peak prices don't seem to cause churn

**Peak Prices:**
- Customers who left pay **9.74% MORE**
- **Conclusion:** Higher peak prices might be pushing customers away!

**Mid-Peak Prices:**
- Customers who left pay **20.14% MORE** 
- **Conclusion:** This is the biggest difference! Mid-peak prices are strongly linked to churn

### **Simple Restaurant Analogy:**

**Think of it like a restaurant with different meal times:**
- **Lunch (off-peak):** Same prices for everyone - no one leaves because of lunch prices
- **Dinner (peak):** Customers who stopped coming were paying 10% more for dinner
- **Brunch (mid-peak):** Customers who stopped coming were paying 20% more for brunch!

### **The Big Insight:**

**Higher prices during busy times (peak & mid-peak) are the main reason customers are leaving!** The company should focus on making peak-time pricing more competitive.

Does this make sense? We're basically proving that expensive "rush hour" prices are driving customers away!

```python
# Visualize the distribution of price variables split by churn
fig, axes = plt.subplots(3, 2, figsize=(15, 18))
fig.suptitle('Price Distribution Analysis by Churn Status', fontsize=16)

# Price columns to analyze (using recent prices)
price_vars = ['recent_price_off_peak', 'recent_price_peak', 'recent_price_mid_peak']

for i, price_var in enumerate(price_vars):
    # Boxplots
    ax1 = axes[i, 0]
    merged_data_complete.boxplot(column=price_var, by='churn', ax=ax1)
    ax1.set_title(f'{price_var} - Boxplot by Churn')
    ax1.set_xlabel('Churn (0=No, 1=Yes)')
    ax1.set_ylabel(price_var)
    
    # Histograms
    ax2 = axes[i, 1]
    churn_0 = merged_data_complete[merged_data_complete['churn'] == 0][price_var].dropna()
    churn_1 = merged_data_complete[merged_data_complete['churn'] == 1][price_var].dropna()
    
    ax2.hist(churn_0, alpha=0.7, label='No Churn', bins=30, color='blue')
    ax2.hist(churn_1, alpha=0.7, label='Churn', bins=30, color='red')
    ax2.set_title(f'{price_var} - Histogram by Churn')
    ax2.set_xlabel(price_var)
    ax2.set_ylabel('Frequency')
    ax2.legend()

plt.tight_layout()
plt.show()
```

## **Step 5: Measure How Connected Prices Are to Customer Leaving**

We're asking: *"How strongly are prices linked to customers leaving? Can we measure this connection?"*

### **What Correlation Means:**

**Think of correlation like measuring friendship:**
- **+1.0** = Perfect positive friendship (when one goes up, the other ALWAYS goes up)
- **0.0** = No friendship at all (they don't affect each other)
- **-1.0** = Perfect opposite friendship (when one goes up, the other ALWAYS goes down)

### **What We Found:**

**üìä The Results (ranked from strongest to weakest):**

1. **Mid-peak prices: +0.046** ‚Üí **Strongest connection!**
2. **Peak prices: +0.030** ‚Üí **Moderate connection**  
3. **Off-peak prices: -0.007** ‚Üí **No real connection**

### **What This Means in Simple Terms:**

**üîç Translation:**
- **Mid-peak prices:** When prices go up, customers are **slightly more likely** to leave
- **Peak prices:** When prices go up, customers are **a little more likely** to leave  
- **Off-peak prices:** Price changes don't really affect who stays or leaves

### **Simple Restaurant Analogy:**

**Think of it like a restaurant:**
- **Brunch prices (mid-peak):** When brunch gets expensive, customers stop coming most often
- **Dinner prices (peak):** When dinner gets expensive, some customers leave
- **Lunch prices (off-peak):** Lunch price changes don't really make people stop coming

### **The Visual Chart:**

The heatmap shows all these connections as colors:
- **Red/Pink** = Positive connection (higher price = more likely to leave)
- **Blue** = Negative connection (higher price = less likely to leave)
- **White** = No connection

### **The Big Insight:**

**Mid-peak pricing has the strongest link to customers leaving!** This confirms our earlier findings - those "rush hour" mid-peak prices are the biggest problem for keeping customers.

Does this make sense? We're basically measuring how much each price type "pushes" customers to leave!

```python
# Correlation analysis between price variables and churn
import numpy as np

price_columns = [
    'recent_price_off_peak', 'recent_price_peak', 'recent_price_mid_peak',
    'avg_price_off_peak', 'avg_price_peak', 'avg_price_mid_peak'
]

correlation_data = merged_data_complete[price_columns + ['churn']].corr()
print("Correlation with Churn:")
print(correlation_data['churn'].sort_values(ascending=False))

# Visualize correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_data, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix: Price Variables and Churn')
plt.tight_layout()
plt.show()
```

```python
# Step 2: Simple analysis for each factor with percentage breakdown
stayed_customers = merged_data_complete[merged_data_complete['churn'] == 0]
left_customers = merged_data_complete[merged_data_complete['churn'] == 1]

numerical_columns = merged_data_complete.select_dtypes(include=['float64', 'int64']).columns
numerical_columns = [col for col in numerical_columns if col not in price_columns and col != 'churn' and col != 'id']

# Store data for visualization
factor_data = []

print("=== DETAILED FACTOR ANALYSIS ===")
for col in numerical_columns:
    stayed_avg = stayed_customers[col].mean()
    left_avg = left_customers[col].mean()
    difference = left_avg - stayed_avg
    
    if stayed_avg != 0:
        percentage_diff = (difference / stayed_avg) * 100
    else:
        percentage_diff = 0
    
    # Calculate percentage within each group (stayed vs left)
    total_stayed = len(stayed_customers)
    total_left = len(left_customers)
    
    print(f"\n{col}:")
    print(f"  Customers who STAYED: {stayed_avg:.2f} (from {total_stayed:,} customers)")
    print(f"  Customers who LEFT:   {left_avg:.2f} (from {total_left:,} customers)")
    print(f"  Difference:           {difference:.2f}")
    print(f"  Percentage difference: {percentage_diff:.2f}%")
    
    # Store for chart
    factor_data.append({
        'factor': col,
        'stayed_avg': stayed_avg,
        'left_avg': left_avg,
        'total_stayed': total_stayed,
        'total_left': total_left,
        'percentage_diff': percentage_diff
    })
```

```python
# Create simplified visual chart showing only the percentage comparison
import matplotlib.pyplot as plt
import numpy as np

# Get top 8 factors with biggest differences
factor_data_sorted = sorted(factor_data, key=lambda x: abs(x['percentage_diff']), reverse=True)[:8]

# Prepare data for plotting
factor_names = [f['factor'].replace('_', ' ').title() for f in factor_data_sorted]
stayed_values = [f['stayed_avg'] for f in factor_data_sorted]
left_values = [f['left_avg'] for f in factor_data_sorted]
percentage_diffs = [f['percentage_diff'] for f in factor_data_sorted]

# Create single chart showing percentage differences
fig, ax = plt.subplots(figsize=(12, 8))

# Create colors based on whether customers who left had higher or lower values
colors = ['red' if diff > 0 else 'green' for diff in percentage_diffs]

bars = ax.barh(factor_names, percentage_diffs, color=colors, alpha=0.7)

# Add labels
ax.set_xlabel('Percentage Difference (%)')
ax.set_title('Top Factors Causing Customer Churn\n(Red = Higher for customers who left, Green = Lower for customers who left)')
ax.axvline(x=0, color='black', linestyle='-', linewidth=0.5)

# Add value labels on bars
for i, (bar, value) in enumerate(zip(bars, percentage_diffs)):
    if value > 0:
        ax.text(value + 1, bar.get_y() + bar.get_height()/2, f'{value:.1f}%', 
                va='center', ha='left', fontweight='bold')
    else:
        ax.text(value - 1, bar.get_y() + bar.get_height()/2, f'{value:.1f}%', 
                va='center', ha='right', fontweight='bold')

# Add explanation text
textstr = f'''
Total Customers Analyzed:
‚Ä¢ Stayed: {factor_data_sorted[0]['total_stayed']:,} customers
‚Ä¢ Left: {factor_data_sorted[0]['total_left']:,} customers
'''
props = dict(boxstyle='round', facecolor='lightblue', alpha=0.8)
ax.text(0.02, 0.98, textstr, transform=ax.transAxes, fontsize=10,
        verticalalignment='top', bbox=props)

plt.tight_layout()
plt.show()

# Print summary of insights
print(f"\n=== KEY INSIGHTS ===")
print(f"üìä TOP 5 FACTORS WITH BIGGEST DIFFERENCES:")
for i, factor in enumerate(factor_data_sorted[:5], 1):
    direction = "HIGHER" if factor['percentage_diff'] > 0 else "LOWER"
    print(f"{i}. {factor['factor']}: Customers who left had {direction} values ({factor['percentage_diff']:.1f}% difference)")
    print(f"   - Stayed group average: {factor['stayed_avg']:.1f}")
    print(f"   - Left group average:   {factor['left_avg']:.1f}")
```

```python
# Step 4: Analyze categorical factors (like channel, origin, etc.)
print("\n=== CATEGORICAL FACTORS ===")

categorical_columns = merged_data_complete.select_dtypes(include=['object']).columns
categorical_columns = [col for col in categorical_columns if col != 'id']

for col in categorical_columns:
    print(f"\n{col.upper()}:")
    
    # Calculate churn rate for each category
    churn_by_category = merged_data_complete.groupby(col)['churn'].agg(['count', 'sum', 'mean']).round(3)
    churn_by_category.columns = ['Total_Customers', 'Customers_Left', 'Churn_Rate']
    churn_by_category['Churn_Percentage'] = (churn_by_category['Churn_Rate'] * 100).round(1)
    
    print(churn_by_category)
```

```python
# Create visual charts for categorical factors - focusing on CHANNEL_SALES
import matplotlib.pyplot as plt

# First, let's create a chart specifically for channel_sales (or similar column)
# Check what categorical columns we have
print("Available categorical columns:", categorical_columns)

# Create charts for each categorical factor
for col in categorical_columns:
    if 'channel' in col.lower() or 'sales' in col.lower() or col in ['channel_sales', 'origin', 'activity_new']:
        
        # Calculate churn data
        churn_data = merged_data_complete.groupby(col)['churn'].agg(['count', 'sum', 'mean']).reset_index()
        churn_data.columns = [col, 'Total_Customers', 'Customers_Left', 'Churn_Rate']
        churn_data['Churn_Percentage'] = (churn_data['Churn_Rate'] * 100).round(1)
        churn_data['Customers_Stayed'] = churn_data['Total_Customers'] - churn_data['Customers_Left']
        
        # Create the chart
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
        
        # Chart 1: Stacked bar chart showing stayed vs left
        categories = churn_data[col]
        stayed = churn_data['Customers_Stayed']
        left = churn_data['Customers_Left']
        
        ax1.bar(categories, stayed, label='Customers Who STAYED', color='green', alpha=0.7)
        ax1.bar(categories, left, bottom=stayed, label='Customers Who LEFT', color='red', alpha=0.7)
        
        ax1.set_title(f'Customer Distribution by {col.title()}')
        ax1.set_xlabel(col.title())
        ax1.set_ylabel('Number of Customers')
        ax1.legend()
        ax1.tick_params(axis='x', rotation=45)
        
        # Add value labels
        for i, (stay, leave) in enumerate(zip(stayed, left)):
            ax1.text(i, stay/2, f'{stay:,}', ha='center', va='center', fontweight='bold')
            ax1.text(i, stay + leave/2, f'{leave:,}', ha='center', va='center', fontweight='bold', color='white')
        
        # Chart 2: Churn rate percentage
        churn_percentages = churn_data['Churn_Percentage']
        colors = ['red' if x > 10 else 'orange' if x > 5 else 'green' for x in churn_percentages]
        
        bars = ax2.bar(categories, churn_percentages, color=colors, alpha=0.7)
        ax2.set_title(f'Churn Rate by {col.title()}')
        ax2.set_xlabel(col.title())
        ax2.set_ylabel('Churn Rate (%)')
        ax2.tick_params(axis='x', rotation=45)
        
        # Add percentage labels on bars
        for bar, pct in zip(bars, churn_percentages):
            height = bar.get_height()
            ax2.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                    f'{pct:.1f}%', ha='center', va='bottom', fontweight='bold')
        
        plt.tight_layout()
        plt.show()
        
        # Print summary
        print(f"\n=== {col.upper()} ANALYSIS ===")
        print(churn_data[[col, 'Total_Customers', 'Customers_Left', 'Churn_Percentage']])
        print(f"Highest churn rate: {churn_data['Churn_Percentage'].max():.1f}% in {churn_data.loc[churn_data['Churn_Percentage'].idxmax(), col]}")
        print(f"Lowest churn rate: {churn_data['Churn_Percentage'].min():.1f}% in {churn_data.loc[churn_data['Churn_Percentage'].idxmin(), col]}")
```
